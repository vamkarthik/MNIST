# Data Loading

- Fetching the data from OpenML using the `fetch_openml` function of the `sklearn.datasets` module.
- The data is returned as a **Bunch object**.
- The `as_frame` parameter is set to `True` to return the data as a **pandas DataFrame**.
- One of the `data_id` or `name` parameters is **required** to fetch the data.
- The `data_id` can be found in the **OpenML website URL** of the dataset.
- The URL for the **MNIST dataset** is [https://www.openml.org/d/554](https://www.openml.org/d/554), indicating the `data_id` is **554**.

```python
from sklearn.datasets import fetch_openml

mnist = fetch_openml(data_id=554,as_frame=True)
```

```python
mnist.keys()
```

```python
from IPython.display import Markdown
display(Markdown(mnist.DESCR))
```

```python
X,y = mnist['data'],mnist['target']
```

```python
import random
import matplotlib.pyplot as plt

idx = random.sample(range(0,len(X)),1)
sample_digit = X.iloc[idx]
label = y.iloc[idx]
sample_digit_image = sample_digit.values.reshape(28,28)
print("The label for this image is {label}".format(label=label.values[0]))
plt.imshow(sample_digit_image,cmap='binary')
plt.show()
```

```python
fig, axes = plt.subplots(10, 10, figsize=(10, 10))

for i in range(100):
    idx = random.sample(range(0, len(X)), 1)
    sample_digit = X.iloc[idx]
    label = y.iloc[idx]
    sample_digit_image = sample_digit.values.reshape(28, 28)

    row = i // 10
    col = i % 10
    axes[row, col].imshow(sample_digit_image, cmap='binary')
    axes[row, col].annotate(f"Label: {label.values[0]}", 
                            xy=(0.5, 0),  # Position below the image
                            xycoords='axes fraction', 
                            ha='center', va='top', fontsize=10)
    axes[row, col].axis('off')

plt.tight_layout()
plt.show()
```
